```python
%load_ext autoreload
%autoreload
from mit_data import MITDataset
from mit_data import train_val_split
from mit_data import make_label_binarizer
import pandas as pd
import numpy as np
```

```python
original_df = pd.read_csv("data/MIT_data/train_index.csv")
list_original = original_df.to_dict("records")
for data in list_original:
    data["object_label_list"] = str(data["object_label"]).split(" ")
```

```python
list_with_man = list(filter(lambda data: "man" in data["object_label_list"], list_original))
list_without_man = list(filter(lambda data: "man" not in data["object_label_list"],list_original))
list_without_man_cvt = list_without_man[:len(list_with_man)]
for data in list_with_man:
    data["object_label"] = "man"
for data in list_without_man_cvt:
    data["object_label"] = ""
```

```python
combined_list = list_with_man + list_without_man_cvt
combined_df = pd.DataFrame(combined_list)
combined_df = combined_df.drop("object_label_list", axis=1)
```

```python
combined_df.to_csv("data/MIT_data/binary_label_man.csv")
```

```python
train_val_split(len(combined_list),save_file="binary_split.csv")
mlb = make_label_binarizer("data/MIT_data/binary_label_man.csv")
```

```python
train_binary_dataset = MITDataset(mlb,mode="train",split_file="binary_split.csv",index_file="binary_label_man.csv")
```
